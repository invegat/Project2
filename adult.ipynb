```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./adult_train.csv")
```

```python
df.head()
```

```python
df.drop(["fnlwgt","education","relationship"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.workclass.isnull().sum()
```

```python
df["native.country"] =  np.where(df["native.country"] != "United-States","other", "United-States")
```

```python
df.tail()
```

```python
df.loc[df.workclass.isin(["Federal-gov", "Local-gov", "State-gov"]), "workclass"] = "Gov"
```

```python
df.loc[df.workclass.isin(["Without-pay", "Neverworked"]), "workclass"] = "Not-working"
```

```python
df.loc[df.workclass.isin(["Self-emp-inc", "Self-emp-not-inc"]), "workclass"] = "Self-employed"
```

```python
df.tail()
```

```python
df.loc[df["marital.status"].isin(["Married-AF-spouse", "Married-civ-spouse"]), "marital.status"] = "Married"
```

```python
df = df.rename(columns={"marital.status": "marital-status", "native.country": "native country"})
```

```python
df.head()
```

```python
df.loc[df["marital-status"].isin(["Married-spouse-absent", "Separated", "Divorced", "Widowed"]), "marital-status"]\
= "Not-married"
```

```python
df.loc[df.occupation.isin(["Tech-support", "Adm-clerical", "Privhouse-serv", "Protective-serv", "Armed-Forces",\
    "Other-service"]),"occupation"] = "Other"
```

```python
df.loc[df.occupation.isin(["Craft-repair","Farming-fishing", "Handlers-cleaners", "Machine-op-inspct", \
                           "Transport-moving"]), "occupation"] = "ManualWork"
```

```python
df = df[df.occupation != '?']
```
